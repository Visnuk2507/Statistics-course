## 2.1 Random Variable (stochastic variable)

### Definition 2.1: Sample Space
- The sample space \( S \) is the set of all possible outcomes of an experiment.

**Example 2.2**: 
- Throwing two paper balls toward a basket:
    - Sample space \( S = \{(miss, miss), (hit, miss), (miss, hit), (hit, hit)\} \)

### Definition 2.3: Random Variable
- A random variable is a function assigning a numerical value to each outcome in the sample space.


**Example 2.4**: 
- Defining a random variable for hits in the paper ball example:
    - $$ X((miss, miss)) = 0 $$
    - $$ X((hit, miss)) = 1 $$
    - $$ X((miss, hit)) = 1 $$
    - $$ X((hit, hit)) = 2 $$

### Probability Distribution of a Random Variable
- Discrete outcomes: dice rolls, number of children, machine failures.
- Continuous outcomes: weight, time spent on homework, electricity generation.

## 2.2 Discrete Random Variables

- A discrete random variable has a set of discrete outcomes and follows a discrete distribution.

### Definition 2.6: Probability Density Function (PDF), alternativt for diskrete variable: pmf
- For a discrete random variable \( X \), the probability density function (pdf) is:
  $$ f(x) = P(X = x) $$
  - $$ f(x) \geq 0, \text{for all x}$$
  - $$ \sum_{all \, x} f(x) = 1 $$



Kan læses som sansynligheden for at X antager værdien x, når eksperimentet udføres

**Example 2.7**:
- PDF of a fair dice roll: $$ P(X = x) = \frac{1}{6} $$ for \( x = 1, 2, 3, 4, 5, 6 \).

### Definition 2.9: Cumulative Distribution Function (CDF)
- The CDF is the probability of realizing an outcome below or equal to \( x \):
  $$ F(x) = P(X \leq x) = \sum_{x_j \leq x} P(X = x_j) $$

### 2.2.1 Introduction to Simulation

- Simulations allow experiments to be conducted virtually using models. Simulations with random numbers are called stochastic simulations.

#### Example 2.11: Simulating Dice Rolls



```python
# Make a random draw from (1,2,3,4,5,6) with equal probability for each 
# outcome
import numpy as np

random_draw = np.random.choice(range(1,7), size = 1)
print(random_draw)
```

The simulation becomes more interesting when the experiment is repeated many
times, then we have a sample and can calculate the empirical density function (or empirical pdf or density histogram, see Section 1.6.1) as a discrete histogram and actually
“see” the shape of the pdf

```python
import matplotlib.pyplot as plt
# Simulate a fair dice
# Number of simulated realizations
n = 30

# Draw independently from the set (1,2,3,4,5,6) with equal probability
xFair = np.random.choice(range(1, 7), size=n, replace=True)

# Count the number of each outcome using the bincount function
counts = np.bincount(xFair)

# Plot the pdf
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,4))

# plt.bar(x, height, width, bottom, align) .  - plotting pdf
ax1.bar(range(1,7), [1/6]*6, color='red')

# Plot the empirical pdf
ax1.bar(range(1,7), counts[1:7]/n)

# Plot the cdf
ax2.bar(range(1,7), np.cumsum([1/6]*6), color='red')

# Add the empirical cdf
ax2.bar(range(1,7), np.cumsum(counts[1:7]/n))


```
