## 1) Diabetes Mellitus – Core Framework

### A. Classification
- **Type 1 DM (T1DM):** Autoimmune destruction of pancreatic β‑cells → absolute insulin deficiency.
- **Type 2 DM (T2DM):** Combination of insulin resistance and β‑cell dysfunction → relative insulin deficiency.
- **Other specific types:**
  - **LADA** (Latent Autoimmune Diabetes in Adults): adult‑onset autoimmune diabetes, slower course than T1DM.
  - **MODY** (Maturity‑Onset Diabetes of the Young): monogenic defects of β‑cell function.
  - **Secondary causes:** pancreatic diseases (pancreatitis, trauma), endocrinopathies, drugs.
  - **Gestational diabetes (GDM):** first recognized during pregnancy.

### B. Etiology
- **T1DM:** genetic predisposition (HLA associations) + environmental trigger (often viral) → autoimmune β‑cell destruction; autoantibodies (ICA, IAA, GAD, etc.).
- **T2DM:** genetic susceptibility + environment → **insulin resistance** (liver, muscle, adipose) + **β‑cell failure** (impaired glucose‑stimulated insulin secretion). Strongly linked to **obesity**, physical inactivity, aging.

### C. Diagnosis (SI units)
- **Fasting plasma glucose (FPG) ≥ 7.0 mmol/L** (after ≥ 8 h fast), **or**
- **Random plasma glucose ≥ 11.1 mmol/L** with classic symptoms, **or**
- **2‑h plasma glucose ≥ 11.1 mmol/L** during 75 g OGTT, **or**
- **HbA1c ≥ 48 mmol/mol**.

### D. Clinical Manifestations
- **Hyperglycemia symptoms:** polyuria, polydipsia, polyphagia, weight loss, fatigue, blurred vision, infections (e.g., candidiasis), poor wound healing.
- **T1DM:** rapid onset; may present with **DKA**.
- **T2DM:** insidious; often detected via complications (e.g., retinopathy) or routine screening.

### E. Pathophysiology (high‑yield)
- **T1DM:**
  - Autoimmune destruction → >80–90% β‑cell loss before symptoms.
  - Near‑total insulin deficiency → unopposed glucagon; ↑ hepatic glucose output; ↑ lipolysis → ketosis.
  - “Honeymoon phase” early after diagnosis (transient partial remission).
- **T2DM:**
  - **Insulin resistance:** ↓ muscle glucose uptake (GLUT4), ↑ hepatic gluconeogenesis/glycogenolysis, adipose insulin resistance → ↑ lipolysis, ↑ FFA flux → worsens IR.
  - **β‑cell dysfunction:** impaired first‑phase insulin response; glucotoxicity & lipotoxicity accelerate decline.
  - **Incretin defect:** blunted GLP‑1 response; obesity associated with ~20% lower GLP‑1.

### F. Complications
- **Acute:**
  - **DKA (T1> T2):** hyperglycemia, ketosis, anion‑gap metabolic acidosis, dehydration.
  - **Hypoglycemia (both types; especially with insulin/SUs):** adrenergic + neuroglycopenic symptoms; can cause seizures/coma.
- **Chronic:**
  - **Microvascular:** retinopathy, nephropathy (albuminuria → CKD), neuropathies (sensory, autonomic).
  - **Macrovascular:** atherosclerotic CVD (MI, stroke, PAD). ~75% of T2DM mortality is CVD.
  - **Others:** infections, foot ulcers/Charcot arthropathy, erectile dysfunction, gastroparesis.

### G. Management (overview)
- **Lifestyle:** medical nutrition therapy, weight loss (5–10%+), regular aerobic + resistance exercise, smoking cessation, sleep, psychosocial support.
- **Pharmacotherapy (T2DM focus):**
  - **Metformin** (first‑line unless contraindicated): ↓ hepatic gluconeogenesis, modest weight loss/neutral.
  - **GLP‑1 receptor agonists** (liraglutide, semaglutide; GLP‑1/GIP duals like tirzepatide): ↓ appetite, slow gastric emptying, ↑ insulin (glucose‑dependent), ↓ glucagon → weight loss (often 10–15%+), HbA1c reduction, CV benefit (agent‑specific).
  - **SGLT2 inhibitors:** glucosuria → ↓ glucose & weight; strong HF/CKD benefits.
  - **DPP‑4 inhibitors:** ↑ endogenous incretins (weight neutral).
  - **Sulfonylureas:** ↑ insulin secretion (risk: hypoglycemia, weight gain).
  - **TZDs:** ↑ insulin sensitivity (risks: weight gain, edema, fractures).
  - **Insulin:** required in **T1DM**; in **T2DM** when oral/GLP‑1 inadequate or acutely ill.
- **CVD risk reduction:** BP control, statins, antiplatelet when indicated, weight management, renal protection.
- **Foot/eye/kidney surveillance:** annual retinal exam, urine albumin/creatinine, monofilament testing.

## 2) Diabetic Ketoacidosis (DKA) – Describe

**Definition:** Acute, life‑threatening complication due to **absolute/relative insulin deficiency** + **excess counter‑regulatory hormones** → uncontrolled lipolysis and ketogenesis.

**Triggers:** infection, missed insulin, new‑onset T1DM, MI, stroke, pancreatitis, steroids, SGLT2i (euDKA risk), dehydration.

**Pathophysiology:**
- Insulin deficiency + high glucagon/epinephrine/cortisol/GH → ↑ hepatic glucose output (gluconeogenesis, glycogenolysis) + ↓ peripheral uptake → **hyperglycemia**.
- Adipose lipolysis → ↑ FFAs → hepatic β‑oxidation → **ketone bodies** (acetoacetate, β‑hydroxybutyrate) → **metabolic acidosis**.
- Osmotic diuresis → **dehydration**, **electrolyte loss** (K⁺ total body deficit despite normal/high serum K⁺ initially).

**Clinical:** polyuria, polydipsia, N/V, abdominal pain, Kussmaul respirations, fruity breath, dehydration, altered mental status.

**Labs:** glucose elevated (can be modest in euDKA), anion‑gap acidosis (↓ HCO₃⁻), ketonemia/ketonuria, variable K⁺.

**Management (principles):**
1) **Fluids first** (e.g., isotonic saline; adjust by hemodynamics/Na⁺).  
2) **Insulin infusion** (after K⁺ ≥ 3.3 mmol/L) to stop ketogenesis.  
3) **Potassium replacement** (anticipate total body deficit).  
4) **Treat precipitant** (e.g., antibiotics for infection).  
5) Add **dextrose** when glucose ~10–14 mmol/L to continue insulin until anion gap closes.

**Complications to watch:** cerebral edema (esp. youth), hypoglycemia, hypokalemia, ARDS.

## 3) Hypoglycemia – Dangers

**Definition (practical):** plasma glucose low enough to cause symptoms/signs; counter‑regulation begins < ~4.0 mmol/L; symptomatic often < ~3.5 mmol/L (varies).

**Symptoms:**
- **Autonomic (adrenergic):** tremor, palpitations, sweating, pallor, hunger, anxiety, nausea.
- **Neuroglycopenic:** confusion, irritability, blurred/double vision, dysarthria, dizziness, seizures, coma.

**Dangers:**
- **Immediate:** accidents, falls, seizures, coma, arrhythmias (QT prolongation), sudden death ("dead‑in‑bed" in T1DM).
- **Impaired awareness:** repeated hypos blunt adrenergic warning symptoms → higher risk of severe events.
- **Cognitive effects:** acute dysfunction; in children/elderly—potential long‑term cognitive impact.
- **Social/functional:** driving/work hazards; fear of hypoglycemia → suboptimal glycemic control.

**Prevention/Management:** education, SMBG/CGM, individualized targets, de‑escalate sulfonylureas/insulin when appropriate, adjust for exercise/alcohol, bedtime snacks if needed, glucagon rescue, rule out contributing adrenal/thyroid dysfunction if recurrent.

## 4) Obesity – Classification, Etiology, Pathophysiology, Risks, Management

### A. Classification
- **By BMI (kg/m²):**
  - Overweight: 25–29.9  
  - Obesity class I: 30–34.9  
  - Obesity class II: 35–39.9  
  - Obesity class III: ≥ 40
- **By fat distribution:** central/visceral vs. peripheral (waist circumference, waist‑to‑hip ratio).

### B. Etiology (multifactorial)
- **Environment/behavior:** caloric surplus, ultra‑processed foods, sedentary lifestyle, short sleep, stress.
- **Biology:** genetic susceptibility; neurohormonal regulation of appetite/satiety (hypothalamus, leptin, GLP‑1, ghrelin, PYY); adaptive thermogenesis.
- **Life course:** early‑onset obesity tracks into adulthood (children obese at age 2 rarely remit without intervention).

### C. Pathophysiology
- **Energy balance:** intake > expenditure.
- **Adaptive responses after weight loss:** disproportionate ↓ resting energy expenditure, ↑ skeletal muscle efficiency, ↑ hunger/cravings, ↑ sedentary time → **weight regain** common.
- **GLP‑1 biology:** obese/overweight individuals exhibit **~20% lower GLP‑1 secretion** post‑meal (independent of glucose tolerance); contributes to impaired satiety.
- **Insulin resistance:** excess visceral fat → adipokine dysregulation, lipotoxicity, inflammation → hepatic/muscle IR.

### D. Risks/Complications
- **Metabolic:** T2DM, metabolic syndrome, NAFLD/NASH, dyslipidemia, hypertension.
- **Cardiovascular:** CAD, stroke, HF; **CVD is the leading cause of death** in T2DM.
- **Other:** OSA, osteoarthritis, GERD, gallstones, subfertility, certain cancers, depression, stigma/reduced QoL.

### E. Management
- **Lifestyle foundation:** calorie deficit (individualized pattern), high‑protein/high‑fiber emphasis, structured physical activity (aerobic + resistance), sleep hygiene, behavioral therapy.
- **Pharmacotherapy:** when BMI ≥ 30 or ≥ 27 with comorbidities.
  - **GLP‑1 RAs / dual incretins (tirzepatide):** large **12–17%** mean weight loss; also improve glycemia and CV risk factors.
  - **Orlistat:** modest additional loss; GI side‑effects.
- **Bariatric/metabolic surgery:** for class II–III obesity or class I with refractory comorbidities; most effective for long‑term weight and T2DM remission.
- **Maintenance strategies:** continued exercise, ongoing pharmacotherapy (if tolerated), frequent follow‑up, relapse planning, address sedentary drift and eating behavior after weight loss.

## 5) Metabolic Syndrome – Describe

**Concept:** Cluster of interrelated risk factors driven by **insulin resistance** and **visceral adiposity** → elevates risk for T2DM & atherosclerotic CVD.

**Typical diagnostic elements (know qualitatively):**
- Central obesity (increased waist circumference, population‑specific cutoffs).
- Elevated triglycerides (≈ ≥ 1.7 mmol/L) and/or low HDL‑C (≈ <1.03 mmol/L men, <1.29 mmol/L women).
- Elevated blood pressure (≈ ≥ 130/85 mmHg or treated).
- Elevated fasting glucose (≈ ≥ 5.6 mmol/L) or established T2DM.

**Pathophysiology:** visceral fat → FFAs, adipokines, inflammation → **insulin resistance**, endothelial dysfunction, pro‑thrombotic state.

**Risks:** markedly ↑ CVD events and progression to T2DM.

**Management:** comprehensive risk reduction—weight loss (5–10%+), physical activity, diet quality (Mediterranean‑like), treat lipids (statins), BP control, stop smoking, consider GLP‑1 RA/SGLT2i when T2DM.

## 6) Glucose‑Regulating Hormones – High‑Yield Profiles

### A. **Insulin**
- **Source:** pancreatic β‑cells.
- **Stimuli:** ↑ plasma glucose (biphasic secretion), incretins (GLP‑1/GIP), amino acids, vagal tone.
- **Receptor/signaling:** receptor tyrosine kinase → PI3K/AKT; translocation of **GLUT4** in muscle/adipose.
- **Major actions:**
  - **Liver:** ↓ gluconeogenesis & glycogenolysis; ↑ glycogenesis; ↑ lipogenesis.
  - **Muscle:** ↑ glucose uptake/oxidation; ↑ glycogen & protein synthesis.
  - **Adipose:** ↑ glucose uptake; ↑ lipogenesis; **↓ lipolysis** (↓ HSL activity).
- **Pathophysiology links:** deficiency (T1DM) vs. resistance (T2DM). Hyperinsulinemia early in T2DM compensates for IR → β‑cell exhaustion.

### B. **Glucagon**
- **Source:** pancreatic α‑cells.
- **Stimuli:** ↓ glucose, amino acids (post‑protein meal), sympathetic activation.
- **Actions:** **Liver‑centric**—↑ glycogenolysis & gluconeogenesis → ↑ plasma glucose; promotes ketogenesis in insulin deficiency.
- **Pathophysiology:** in diabetes, glucagon is **inappropriately elevated**, exacerbating fasting/post‑prandial hyperglycemia.

### C. **Incretins / GLP‑1**
- **Source:** intestinal L‑cells (jejunum/ileum); released with meals.
- **Actions:** glucose‑dependent **↑ insulin**, **↓ glucagon**, **slows gastric emptying**, **↑ satiety** (CNS/vagal pathways), supports β‑cell function.
- **Clinical:** pharmacologic **GLP‑1 RAs** → weight loss, HbA1c reduction, some agents with CV benefit; obesity shows **blunted endogenous GLP‑1** responses.

### D. **Growth Hormone (GH)**
- **Source:** anterior pituitary; pulsatile; ↑ with sleep, stress, exercise, hypoglycemia.
- **Metabolic effects:** **anti‑insulin** in periphery—↑ lipolysis, ↓ glucose uptake in muscle/adipose; **↑ hepatic IGF‑1**; overall can raise fasting glucose and contribute to IR.

### E. **Epinephrine (Adrenaline)**
- **Source:** adrenal medulla; sympathetic activation.
- **Effects:** **counter‑regulatory**—↑ hepatic glycogenolysis & gluconeogenesis; **↓ insulin secretion**, **↑ glucagon**; **↑ lipolysis**; acutely raises plasma glucose; vital in hypoglycemia defense.

## 7) MODY – Mention & Define

**Maturity‑Onset Diabetes of the Young (MODY):**
- **Definition:** Monogenic, autosomal‑dominant forms of diabetes due to primary **β‑cell dysfunction**, typically **before age 25**, non‑ketotic, non‑obese, preserved insulin sensitivity.
- **Common genes/subtypes:** **GCK (MODY2)**—mild stable fasting hyperglycemia; **HNF1A (MODY3)** and **HNF4A (MODY1)**—progressive; **HNF1B (MODY5)** with renal anomalies.
- **Clinical clues:** multigenerational inheritance, negative autoantibodies, low insulin requirements, discordant phenotype for T1/T2.
- **Management:**
  - **GCK‑MODY:** usually **no pharmacotherapy** (benign course; treat in pregnancy if needed).
  - **HNF1A/HNF4A‑MODY:** high **sulfonylurea sensitivity**; may avoid insulin.
  - Genetic testing confirms and guides therapy.

## 8) Putting It Together – Comparison Tables (High‑Yield)

### A. **Type 1 vs. Type 2 Diabetes**

| Feature | Type 1 DM | Type 2 DM |
|---|---|---|
| Onset | Usually childhood/adolescence; can be adult | Usually adult; rising in youth |
| Etiology | Autoimmune β‑cell destruction | Insulin resistance + β‑cell dysfunction |
| Body habitus | Lean/any | Overweight/obese common |
| Autoantibodies | Present (ICA, GAD, IAA, etc.) | Absent |
| Insulin levels | Very low/undetectable | Normal/high early → low later |
| Ketosis | Common at presentation | Less common (but possible) |
| Treatment | **Insulin required** | Lifestyle ± oral/GLP‑1/SGLT2; insulin if needed |

### B. **Hypoglycemia vs. Hyperglycemia – Quick Contrast**

| Aspect | Hypoglycemia | Hyperglycemia |
|---|---|---|
| Key symptoms | Sweating, tremor, palpitations; confusion, seizures | Polyuria, polydipsia, weight loss, fatigue |
| Immediate risks | Accidents, arrhythmias, seizures, coma | DKA/HHS, dehydration |
| Long‑term | Impaired awareness, cognitive effects | Micro/macrovascular complications |
| First aid | Fast carbs (15–20 g), recheck in 15 min; glucagon if severe | Fluids, insulin if indicated; address cause |

### C. **Antihyperglycemic Agents – Very Brief Exam Cheat**

| Class | A1c ↓ | Weight | Hypos | CV/Kidney |
|---|---|---|---|---|
| Metformin | ~1–1.5% | Neutral/↓ | Rare | Possible CV benefit |
| GLP‑1 RA / dual incretin | ~1–2% | **↓↓** | Rare | Some agents ↓ MACE |
| SGLT2i | ~0.5–1% | ↓ | Rare | **HF & CKD benefits** |
| DPP‑4i | ~0.5–0.7% | Neutral | Rare | Neutral |
| SU | ~1–1.5% | ↑ | **Yes** | Neutral |
| Insulin | Unlimited | ↑ | **Yes** | Required in T1DM |

## 9) Slide‑Specific Pearls (from Torekov deck)
- **Global burden:** >500 million with obesity; 1 in 10 has diabetes; obesity linked to a **3‑fold** CVD risk and large healthcare costs.
- **Weight regain:** ~50% within 1 year after weight loss due to ↓ energy expenditure, ↑ hunger, and ↑ sedentary time.
- **GLP‑1 in obesity:** ~20% lower post‑meal GLP‑1 secretion in overweight/obese independent of glucose tolerance.
- **Exercise + GLP‑1 RA:** additive reductions in **body weight** and **fat%**, improved **HbA1c**, **insulin sensitivity**, **cardiorespiratory fitness**, and **QoL**.
- **Real‑life after stopping therapy:** behavioral drift toward increased sedentary time and eating → regain risk; long‑term plans matter.
- **Adolescents with early‑onset obesity:** semaglutide shows benefit (RESETTLE RCT context).
- **Public health:** childhood obesity tracks strongly into adulthood; inequities documented in large Danish cohorts.

## 10) Quick Oral‑Exam Prompts & Model Outlines

1) **“Explain T1DM pathophysiology.”**  
Autoimmune β‑cell destruction (HLA‑linked) → absolute insulin deficiency → ↑ hepatic glucose output + ↓ peripheral uptake; ↑ lipolysis → ketone production → risk of DKA. Early “honeymoon” may transiently lower insulin needs.

2) **“Three key metabolic abnormalities in T2DM.”**  
(1) Insulin resistance (muscle, liver, adipose).  (2) β‑cell dysfunction (impaired first‑phase and progressive failure).  (3) Incretin defect + inappropriate glucagon.

3) **“Describe metabolic syndrome.”**  
Cluster of central obesity, dyslipidemia (↑TG/↓HDL), elevated BP, and impaired fasting glucose—driven by insulin resistance and visceral adiposity—raises CVD and T2DM risk.

4) **“List hypoglycemia symptoms and why it’s dangerous.”**  
Adrenergic (sweat, tremor, palpitations) + neuroglycopenic (confusion, seizures, coma). Dangerous due to accidents, arrhythmias, death; repeated episodes cause impaired awareness.

5) **“Define MODY and give two examples.”**  
Monogenic AD diabetes presenting in youth with β‑cell dysfunction; **GCK‑MODY** (mild stable FPG), **HNF1A‑MODY** (SU‑sensitive).

6) **“DKA management steps.”**  
Fluids → insulin (after K⁺ ≥ 3.3) → K⁺ replacement → treat trigger → dextrose when glucose ~10–14 mmol/L until anion gap closes.

### Final Tips
- Always pair **glucose control** with **CVD risk management** (BP, lipids, smoking).  
- In obesity, plan for **maintenance** from day 1 (exercise + pharmacotherapy often needed).  
- Think **hormone physiology** at each step: insulin vs. counter‑regulators (glucagon, epinephrine, GH) and the **incretin axis**.
