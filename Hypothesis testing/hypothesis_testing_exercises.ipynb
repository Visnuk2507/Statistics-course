### **a)**

### To investigate whether the requirement to the mean is fulfilled (with  = 5%), the following hypothesis should be tested

H0 :  = 3000
H1 :  = 3000.

### Or similarly asked: what is the evidence against the null hypothesis?


### **Solution:**

Its a one-sample sitation we can do, which in python could be handled by:

```python
x = np.array([3003, 3005, 2997, 3006, 2999, 2998, 3007, 3005, 3001])

print(stats.ttest_1samp(x, popmean=3000)) #popmean is the null hypothesis. statistic is the Tobs value

#or
t_obs, p_val = stats.ttest_1samp(x,popmean=3000)
print(t_obs)

print(p_val)


```

### **b)**

### What would the level  = 0.01 critical values be for this test, and what are the interpretation of these?


### **Solution:**

The critical values would be $\pm t_{0.995}=\pm3.355$ 

```python
print(stats.t.ppf(0.995, df = 8))
```

This means that, in a new experiment, the standardized difference between the data
and the null hypothesis, also called tobs, must be either larger than 3.355 or smaller
than 3.355 to lead to a significant result of the experiment.

### **c)**

### What would the level  = 0.05 critical values be for this test (compare also with the values found in the previous question)?

```python
print(stats.t.ppf(0.975, df = 8))
```

This means that, again in a new experiment, it is easier to detect an effect with significance level  = 0.05 than on level  = 0.01.

## **Exercise 3.4** Aluminum profile (hypothesis testing)




### **a)**

### Find the evidence against the following hypothesis: H0 :  = 180

### **Solution:**

```python
x = np.array([180.02, 180.00, 180.01, 179.97, 179.92, 180.05, 179.94,180.10, 180.24, 180.12, 180.13, 180.22, 179.96, 180.10, 179.96, 180.06])

print(stats.ttest_1samp(x, popmean=180)) #popmean is the null hypothesis
```

Therefore weak evidence against H0

### **b)**

### If the following hypothesis test is carried out H0 :  = 180, H1 :  = 180. What are the level  = 1% critical values for this test?


### **Solution:**

The critical values are the 0.005 and 0.995 quantiles of the t-distribution with  = n  1 = 15 degrees of freedom, t0.995 = 2.947:


```python
# 16 (n) - 1 = 15
print(stats.t.ppf(0.995, df = 15))
```

### **c)**

### What is the 99%-confidence interval for ?


### **Solution:**

```python
(CI_low,CI_high) = stats.ttest_1samp(x, popmean=180).confidence_interval(confidence_level=0.99)
print(CI_low,CI_high)
```
