### **Mean and Variance**
For a continuous random variable, the mean (or expected value) is:
$$
\mu = E(X) = \int_{-\infty}^{\infty} x f(x) dx
$$
The variance is:
$$
\sigma^2 = E[(X - \mu)^2] = \int_{-\infty}^{\infty} (x - \mu)^2 f(x) dx
$$
- In continuous cases, integrals are used instead of sums (discrete case).
- Probability of observing a single value is always zero in continuous random variables.

### **2.5.1 Uniform Distribution**

A uniform distribution is a type of probability distribution in which every outcome in a certain range is equally likely. The key parameters for the uniform distribution are:

* α: The lower bound (minimum value) of the range of possible outcomes.

* β: The upper bound (maximum value) of the range of possible outcomes.

In a uniform distribution between α and 𝛽 any value between 
𝛼 and 𝛽 has an equal chance of occurring. 


A uniform random variable \(X\) is distributed between two bounds \(\alpha\) and \(\beta\).

PDF:
$$
f(x) = \begin{cases} \frac{1}{\beta - \alpha} & \text{for } x \in [\alpha, \beta] \\ 0 & \text{otherwise} \end{cases}
$$




CDF:
$$
F(x) = \begin{cases} 0 & \text{for } x < \alpha \\ \frac{x - \alpha}{\beta - \alpha} & \text{for } x \in [\alpha, \beta] \\ 1 & \text{for } x \geq \beta \end{cases}
$$



Mean:
$$
\mu = \frac{\alpha + \beta}{2}
$$
Variance:
$$
\sigma^2 = \frac{(\beta - \alpha)^2}{12}
$$

### **2.5.2 Normal Distribution**

The normal distribution is widely used and appears in many natural phenomena.
PDF:
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$
CDF:
$$
F(x) = \frac{1}{\sigma \sqrt{2\pi}} \int_{-\infty}^{x} e^{-\frac{(u - \mu)^2}{2\sigma^2}} du
$$
Standard normal distribution: \(Z \sim N(0, 1)\).
Transformation to standard normal:
$$
Z = \frac{X - \mu}{\sigma}
$$

### **2.5.3 Log-Normal Distribution**
If \(X\) is log-normal distributed, then \( \ln(X) \) follows a normal distribution.
PDF:
$$
f(x) = \frac{1}{x \sqrt{2 \pi \beta}} e^{-\frac{(\ln x - \alpha)^2}{2 \beta^2}}
$$
Mean:
$$
\mu = e^{\alpha + \beta^2 / 2}
$$
Variance:
$$
\sigma^2 = e^{2\alpha + \beta^2}(e^{\beta^2} - 1)
$$

## **Chapter 2.6:** Simulation of Random Variables
Simulation of random variables uses random number generators (RNGs).
The basic technique involves generating random numbers from the uniform distribution \(U(0, 1)\) and then transforming these to follow a desired distribution.
For the exponential distribution \(X \sim \text{Exp}(\lambda)\):
- If \(U \sim U(0, 1)\), then:
$$
X = -\frac{1}{\lambda} \ln(1 - U)
$$

## **Chapter 2.7:** Identities for the Mean and Variance
- For a linear transformation \(Y = aX + b\):
  - Mean: \( E(Y) = aE(X) + b \)
  - Variance: \( V(Y) = a^2 V(X) \)
- For linear combinations of independent random variables \( Y = a_1 X_1 + a_2 X_2 + \dots + a_n X_n \):
  - Mean:
$$
E(Y) = a_1 E(X_1) + a_2 E(X_2) + \dots + a_n E(X_n)
$$
  - Variance:
$$
V(Y) = a_1^2 V(X_1) + a_2^2 V(X_2) + \dots + a_n^2 V(X_n)
$$

## **Chapter 2.8:** Covariance and Correlation
**Covariance** between two random variables \(X\) and \(Y\) is:
$$
\text{Cov}(X, Y) = E[(X - E[X])(Y - E[Y])]
$$
Covariance of independent variables is zero, but the reverse isn't necessarily true.
**Correlation**:
$$
\rho_{XY} = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
$$
Correlation is a number between -1 and 1, quantifying the linear dependence between two variables.
